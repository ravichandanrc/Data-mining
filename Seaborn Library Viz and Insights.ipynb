### Importing the necessary libraries

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

### Loading the penguin dataset

```python
penguins = sns.load_dataset("penguins")
```

```python
# structure of the dataset
print("***Shape of the Penguin dataset***")
print(penguins.shape)
print("\n")

# sample look at the dataset values
print("***Sample look at the dataset values***")
print(penguins.head())
print("\n")

print("***Dataset info before dropping missing values***")
print(penguins.info())
print("\n")


# dropping the data with missing values
penguins = penguins.dropna()

print("***Shape of the Penguin dataset after removing the rows with missing values***")
print(penguins.shape)
print("\n")

print("***Observe IMPROPER INDEX after dropping missing values***")
print(penguins.head())
print("\n")

# resetting the index
penguins = penguins.reset_index(drop=True)

# checking if the index is reset and for any missing values
print("***Observe INDEX after its reset***")
print(penguins.head())
print("\n")

print("***Dataset info after dropping missing values***")
print(penguins.info())
print("\n")
```

## Dataset Info

After loading the required libraries necessary for our EDA, the penguins dataset has been imported. By checking the shape of the dataset, we can see that it has 344 rows and 7 columns. After looking at the sample data values, we can observe that there are missing values in some records. After checking the info of the dataset, we can deduce that each float datatype column has 2 missing values, and also a number of 11 missing values in the ‘sex’ column.
These missing values are removed and this in turn disturbs the index of the dataset which is reset as specified. After verifying the index reset and confirming that there are no missing values, EDA is carried out to extract any valuable information about the dataset.


## Exploratory Data Analysis

Data Visualization is one of the best steps to have a better understanding of our dataset. The 3 plots and the observations from them are as follows:

```python
# setting the theme
sns.set_theme()
sns.set(rc={"figure.figsize": (8,6), "figure.dpi": 120})

# plot 1: Histogram
sns.histplot(data=penguins, x="flipper_length_mm", hue="species", kde=True).set(title="Plot #1: Histogram showing various penguin species")
plt.show()
```

- We can see the visual representation of the penguins dataset in terms of flipper_length_mm categorical attribute, with a histogram.
- Three groups of penguins species are overlaid for comparing distributions easily.
- We can observe that there are not many outliers in the dataset. 
- The curved line represented the KDE (Kernel density estimation).


```python

# plot 2: Violin plot
sns.violinplot(x="species", y="bill_length_mm", data=penguins, hue="sex",split=True, inner = "quartile").set(title="Plot #2: Violin plot depicting the distribution of data")
plt.show()
```

- A violin plot has been made to distinguish between male and female sex within each species.
- The distribution of numerical variable (bill_length_mm) in terms of the categorical variable (species) can be compared efficiently within this violin plot. 
- We can quickly estimate how the data values of each species are spread. The dotted lines represent first, second and third quartiles. The second quartile is essentially the median.


```python
# plot 3: Heat map
sns.heatmap(penguins.corr(), cmap="YlGnBu", annot=True).set(title="Plot #3: Heat map showing correlations between the variables")
plt.show()
```

- The correlation between the variables on each axis is represented by a square.
- The correlation coefficient, which ranges from +1 to -1, is used to express correlation.
- We can observe that there is a high positive correlation between the flipper_length_mm and body_mass_g variables, which means that their values increase or decrease together. 
- Also, there is a negative correlation between bill_depth_mm and body_mass_g variables, which implies that they tend to move in opposite directions.
- Since the same two variables are being paired together, the plot is symmetrical about the diagonal.
