### Load the Iris dataset

```python
iris = datasets.load_iris()
print("***IRIS SAMPLE DATA***")
print(iris.data[0:5])
print("***SHAPE OF IRIS DATASET***")
print(iris.data.shape)
print(">>> Each row represents the flowers and each column represents the length and width")

print("***IRIS FEATURE NAMES***")
print(iris.feature_names)

print("***IRIS SPECIES***")
print(iris.target_names)

print("***TARGET VALUES***")
print(iris.target)
print(">>> By Observation: 0 = Setosa, 1 = Versicolor, 2 = Virginica")
```

### SCATTER PLOTS

Usually employed with larger data sets, SCATTER PLOTS can be helpful for spotting patterns, relationships, and outliers in data sets.

```python
# The indices of the features that we are plotting
x_index = 0 
y_index = 1

# Basic Scatter plot between two attributes - Sepal length and Sepal width
plt.figure(figsize=(5, 4))
plt.scatter(iris.data[:, x_index], iris.data[:, y_index], c=iris.target)
plt.xlabel(iris.feature_names[x_index])
plt.ylabel(iris.feature_names[y_index])
plt.tight_layout()
plt.title(label='#1. Basic Scatter plot')
```

```python
# formatter to display the appropriate target names on the colorbar
formatter = plt.FuncFormatter(lambda i, *args: iris.target_names[int(i)])

plt.figure(figsize=(5, 4))
plt.scatter(iris.data[:, x_index], iris.data[:, y_index], c=iris.target)
plt.colorbar(ticks=[0, 1, 2], format=formatter) # colour coding the markers by species
plt.xlabel(iris.feature_names[x_index])
plt.ylabel(iris.feature_names[y_index])
plt.tight_layout()
plt.title(label='#2. Scatter plot with colour-coded markers')
```

### Observation from Figure 2:
- Compared to other species, Setosa Sepals have a wider base and a shorter tip.
- Sepal lengths of Virginica are longer than those of Versicolor.

By examining the plot, the Setosa species in Violet can be easily distinguished from Versicolor and Virginica, which are shown in green and yellow colours respectively. However, because they overlap, it is still difficult to tell them apart. We can use petal length and petal width features of the data to get more information.


```python
# Scatter plot between the attributes - Petal length and Petal width

formatter = plt.FuncFormatter(lambda i, *args: iris.target_names[int(i)])
plt.figure(figsize=(5, 4))
plt.scatter(iris.data[:, 2], iris.data[:, 3], c=iris.target)
plt.colorbar(ticks=[0, 1, 2], format=formatter) # colour coding the markers by species
plt.xlabel(iris.feature_names[2])
plt.ylabel(iris.feature_names[3])
plt.tight_layout()
plt.title(label='#3. Scatter plot with colour-coded markers')
plt.show()
```

### Observations from Figure 3:

We can conclude that petal length and petal width attributes are a better combination to distinguish the species Versicolor and Virginica, as there is comparatively a minor overlap between the two.
